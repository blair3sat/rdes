```python
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('ggplot')

from astropy import units as u
import poliastro

from poliastro.bodies import Earth
from poliastro.twobody import Orbit

from poliastro.plotting import plot

# Define our orbit
EARTH_RAD = 6371 * u.km
ALTITUDE = 350 * u.km
ECC = 0 * u.one
INC = 51.6 * u.deg
RAAN = 0 * u.deg
PERICENTER = 0 * u.deg
TANOMALY = 0 * u.deg

sato = Orbit.from_classical(Earth,
                            EARTH_RAD + ALTITUDE,
                            ECC,
                            INC,
                            RAAN,
                            PERICENTER,
                            TANOMALY)

# Plot our orbit
plot(sato)
```
